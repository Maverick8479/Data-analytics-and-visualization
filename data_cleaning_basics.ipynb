```r
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```r
data = pd.read_csv('kc_house_data.csv')
```

```r
data.head()
```

```r
import seaborn as sns
import plotly.express as px
```

```r
plt.figure(figsize=(10, 6))
plt.scatter(data['sqft_living'], data['price'], alpha=0.5)
plt.title('Price vs. Square Footage')
plt.xlabel('Square Footage')
plt.ylabel('Price')
plt.show()
```

```r
sns.pairplot(data[['price', 'sqft_living', 'bedrooms', 'bathrooms']])
plt.show()
```

```r
fig = px.scatter(data, x='sqft_living', y='price', color='bedrooms', size='bathrooms', hover_data=['zipcode'])
fig.update_layout(title='Interactive Price vs. Square Footage', xaxis_title='Square Footage', yaxis_title='Price')
fig.show()
```

```r
bedroom_counts = data['bedrooms'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(bedroom_counts.index, bedroom_counts.values, color='skyblue')
plt.title('Distribution of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Count')
plt.show()
```

```r
grade_counts = data['grade'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(grade_counts, labels=grade_counts.index, autopct='%1.1f%%', startangle=90, colors=['gold', 'lightcoral', 'lightgreen'])
plt.title('Distribution of House Grade')
plt.show()
```

```r
correlation_matrix = data.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

```r
library(ggplot2)
library(lattice)

```

```r
data <- read.csv('kc_house_data.csv')
```

```r
barplot(table(data$bedrooms), col = 'skyblue', main = 'Distribution of Bedrooms', xlab = 'Number of Bedrooms', ylab = 'Count')
```

```r
pie(table(data$grade), labels = paste(names(table(data$grade)), "\n", table(data$grade)), col = c('gold', 'lightcoral', 'lightgreen'), main = 'Distribution of House Grade')
```

```r
ggplot(data, aes(x = sqft_living, y = price)) +
  geom_point(alpha = 0.5) +
  labs(title = 'Price vs. Square Footage', x = 'Square Footage', y = 'Price')
```

```r
ggplot(data, aes(x = as.factor(bedrooms), y = price)) +
  geom_boxplot(fill = 'skyblue') +
  labs(title = 'Price distribution based on Number of Bedrooms', x = 'Number of Bedrooms', y = 'Price')
```

```r
ggplot(data, aes(x = sqft_living, y = price, color = bedrooms)) +
  geom_point(alpha = 0.7, size = 3) +
  scale_color_gradient(low = "blue", high = "red") +
  labs(title = 'Price vs. Square Footage with Bedroom Color Gradient', x = 'Square Footage', y = 'Price')
```

```r
ggplot(data, aes(x = sqft_living, y = price)) +
  geom_point(alpha = 0.7, size = 3) +
  facet_wrap(~bedrooms, scales = "free") +
  labs(title = 'Price vs. Square Footage Faceted by Bedrooms', x = 'Square Footage', y = 'Price')
```
